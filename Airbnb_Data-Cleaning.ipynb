# Checking for Outliers

```python
# Function to check outliers in Column
def outliers_count(col):
    Q1 = col.quantile(.25)
    Q3 = col.quantile(.75)
    IQR = Q3 - Q1
    upper_threshold = Q3 + 1.5*IQR
    lower_threshold = Q1 - 1.5*IQR

    count_of_outliers = 0

    for num in col:
        if num > upper_threshold or num < lower_threshold:
            count_of_outliers += 1

    return count_of_outliers
```

```python
# Checking for outliers in price column
print('Price :', outliers_count(df.price))
```
